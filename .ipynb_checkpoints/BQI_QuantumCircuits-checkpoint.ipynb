```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.quantum_info import Operator
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

The Aer Simulator is used to simulate quantum circuits. We can build the composition of operations THSH on the qubit X as follows:

```python
circuit = QuantumCircuit(1)
circuit.h(0)
circuit.s(0)
circuit.h(0)
circuit.t(0)

display(circuit.draw(output="mpl"))
```

Note that the default names for qubits in Qiskit are $q_0, q_1, q_2$, etc.. If we wish to choose our own name we can use the `QuantumRegister` class, which allows us to name a collection of qubits and treat it as a single object.

```python
X = QuantumRegister(1, "X")
circuit = QuantumCircuit(X)
circuit.h(0)
circuit.s(0)
circuit.h(0)
circuit.t(0)

display(circuit.draw(output="mpl"))
```

Consider a circuit that converts the standard basis to the Bell basis:

```python
X = QuantumRegister(1, "X")
Y = QuantumRegister(1, "Y")
circuit = QuantumCircuit(Y, X)  # note the ordering here 
circuit.h(Y)
circuit.cx(Y, X)                # note the ordering here as well

display(circuit.draw(output="mpl"))
display(Operator.from_circuit(circuit).draw("latex"))
```

Note that qubit wires are represented as single solid lines, wherea classical bit wires will be represented as double solid lines as shown below:

```python
X = QuantumRegister(1, "X")
Y = QuantumRegister(1, "Y")
A = ClassicalRegister(1, "A")
B = ClassicalRegister(1, "B")

circuit = QuantumCircuit(Y, X, B, A)
circuit.h(Y)
circuit.cx(Y, X) 
circuit.measure(Y, B)
circuit.measure(X, A)

display(circuit.draw(output="mpl"))
```

This circuit can be simulated using the Aer Simulator like this:

```python
result = AerSimulator().run(circuit).result()
statistics = result.get_counts()
display(plot_histogram(statistics))
```
